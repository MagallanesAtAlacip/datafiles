# Retrieve the data

Link to file:

```python
wideLink="https://github.com/MagallanesAtAlacip/datafiles/raw/main/covid_fatalities2021_yearly_wide.csv"
```

Reading from link:

```python
import pandas as pd

covidPeru_wide=pd.read_csv(wideLink)
covidPeru_wide.head()
```

```python
covidPeru_wide.info()
```

As you see, each row is a province. Then each column as a specific count. for example:
> The first cell next to ABANCAY tells there were **15** fatalities which were less than 40 in the year 2020.

A functional approach can work well with this:


```python
import matplotlib.pyplot as plt

plt.boxplot(covidPeru_wide.drop(columns=["PROVINCIA"]).values,
            tick_labels=covidPeru_wide.drop(columns=["PROVINCIA"]).columns);
```

Instead of pure matplotlib, we can use pandas wraper:

```python
covidPeru_wide.boxplot(vert=False)
plt.xscale('log')
```

Or seaborn wraper:

```python
import seaborn as sns
sns.boxplot(data=covidPeru_wide, orient='h', log_scale=True, color='skyblue')
```

But, several details will need to call matplotlib mothership!

```python
from matplotlib.ticker import ScalarFormatter

sns.boxplot(data=covidPeru_wide, orient='h', log_scale=True, color='skyblue')

# Matplotlib has to step in
plt.gca().xaxis.set_major_formatter(ScalarFormatter())
plt.ticklabel_format(style='plain', axis='x')
```
